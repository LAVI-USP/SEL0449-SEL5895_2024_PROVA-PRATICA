### 1) Realce - Reconhecimento de digital:

Na atualidade, muitas são as situações nas quais a impressão digital é utilizada como chave para o reconhecimento de indivíduos. Com aplicações desde o desbloqueio de celulares, abertura de portas, e até mesmo a validação de transações bancárias, é crucial, para a segurança e confiabilidade destes processos, que tal reconhecimento seja o mais preciso possível.

Um importante passo no reconhecimento automatizado de impressões digitais é o realce das cristas das impressões.

Neste exercício, utilizaremos, portanto, da filtragem passa-alta para realçar as cristas da impressão digital contida na imagem ```fingerprint.tif```, digitalizada com de **resolução de 684 DPI**.


1. Carregue a imagem ```fingerprint.tif``` e mostre-a.
2. Passe esta imagem para o domínio da frequência (espectro de Fourier). Em seguida, mostre a magnitude do espectro (**Lembre-se do padding**).
3. Crie um filtro de realce (50%) utilizando um passa-alta do tipo “Butterworth” (n = 4) com frequência de corte **1,3 ciclos/mm**. Aplique na imagem e verificar o resultado (mostre a magnitude do filtro criado, a imagem filtrada e o espectro de frequências filtrado).
4. Binarize, através do método de Otsu, a imagem original e a imagem filtrada. Mostre os resultados lado a lado.
5. Comente os resultados.

````
# Seu código começa aqui

# Item 1
fingerprint = cv.imread('fingerprint.tif', cv.IMREAD_GRAYSCALE)

print(fingerprint.shape)
print(fingerprint.dtype)

plt.figure(figsize=(5,5))
plt.imshow(fingerprint,'gray')
plt.title('Imagem original')
plt.show()

# Seu código termina aqui
````

````
# Item 2
M,N = fingerprint.shape
print(M,N)

#padding
Mf = 2*M
Nf = 2*N

f = np.fft.fft2(fingerprint,s=(Mf,Nf))
fshift = np.fft.fftshift(f)
print(fshift.dtype)
magnitude_spectrum = 20*np.log(np.abs(fshift)+1.0)
print(magnitude_spectrum.shape)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(fingerprint,'gray')
plt.title('Imagem original')
plt.subplot(1,2,2)
plt.imshow(magnitude_spectrum,'gray')
plt.title('Espectro - Imagem original')
plt.show()
````

````
# Item 3

c_u = Mf//2
c_v = Nf//2


print("Centro: ", c_u, c_v)

# 684 pixels --- 1 polegada = 25,4 [mm]
#   1 pixel  --- delta_x [mm]

delta_x = 25.4/684
print("delta_x =", delta_x)

u_max = 1/(2*delta_x)              # freq. de Nyquist em ciclos/mm
print("u_max =", u_max)
# M pixels --- u_max [ciclos/mm]
#   1 pixel  --- delta_u [ciclos/mm]
delta_u = u_max/(M)              # em ciclos/mm
print("delta_u =", delta_u)

#  1 pixel  --- delta_u [ciclos/mm]
# D0 pixels --- 1,2 [ciclos/mm]

D0 = 1.3//delta_u

print("D0 =", D0)


butter_n = 4
filtro_pa_butterworth = np.zeros((Mf,Nf))
for u in range(0,Mf):
    for v in range (0,Nf):
        D = np.sqrt((u-c_u)**2+(v-c_v)**2)
        filtro_pa_butterworth [u,v] = (1 - 1/(1+(D/D0)**(2*butter_n)))*0.5 + 1

mag_filtro_pa_butterworth = fshift * filtro_pa_butterworth

ifshift = np.fft.ifftshift(mag_filtro_pa_butterworth)
img_filtro_pa_butterworth = np.fft.ifft2(ifshift)
img_filtro_pa_butterworth_abs = np.abs(img_filtro_pa_butterworth)
img_filtro_pa_butterworth = img_filtro_pa_butterworth_abs[0:M,0:N].astype('uint8')

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(fingerprint,'gray')
plt.title('Imagem original')
plt.subplot(2,2,2)
plt.imshow(filtro_pa_butterworth, 'gray', vmin = 0, vmax = 1.5)
plt.title('Filtro de Realce')
plt.subplot(2,2,3)
plt.imshow(img_filtro_pa_butterworth,'gray')
plt.title('Imagem filtrada')
plt.subplot(2,2,4)
plt.imshow(20*np.log(np.abs(mag_filtro_pa_butterworth)+1.0),'gray')
plt.title('Espectro filtrado')
plt.show()
````

````
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(fingerprint,'gray')
plt.title('Imagem original')
plt.subplot(1,2,2)
plt.imshow(img_filtro_pa_butterworth, 'gray')
plt.title('Imagem filtrada')
plt.show()
````

````
# Item 4
ret2,original_binarizada = cv.threshold(fingerprint,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
ret2,filtrada_binarizada = cv.threshold(img_filtro_pa_butterworth,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(original_binarizada, cmap='gray', vmin=0, vmax=255)
plt.title("Original binarizado")
plt.subplot(1,2,2)
plt.imshow(filtrada_binarizada, cmap='gray', vmin=0, vmax=255)
plt.title("Filtrada binarizado")
plt.show()
````
